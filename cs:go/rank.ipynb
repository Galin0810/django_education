```python
#res_df['date'] = pd.to_datetime(res_df['date'])
#res_df = results_df[res_df['date'] > "31-12-2018"]

def probability(rating1, rating2): 
    
    return (1.0 * 1.0 / (1 + 1.0 * pow(10, 1.0 * (rating1 - rating2) / 400)) )
    
 
def elo_rating(R_T_1, R_T_2, K, s): 
    df1 = pd.DataFrame(
        columns = [ 'R_T_1', 'R_T_2', 'team_1', 'team_2', 'date'])
        
    RT1=0
    RT2=0
    results_df.insert(5, 'R1', 0)
    results_df.insert(5, 'R2', 0)

    Pb = probability(R_T_1, R_T_2) 
  
    Pa = probability(R_T_1, R_T_2) 
    #results_df.loc[results_df.match_winner == 1,'R1'] = 100
    
    #results_df.loc[results_df.match_winner == 2, 'R2'] = 100
    
    res_df['R1'] = res_df.map_winner.map(lambda x: 100 if x == 1 else 0)  
    res_df['R2'] = res_df.map_winner.map(lambda x: 100 if x == 2 else 0) 
    res_df['R1'] = res_df.major.map(lambda x: 1600 if x == True else 0)
    res_df['R2'] = res_df.major.map(lambda x: 1600 if x == True else 0)
    res_df['R1'] = res_df.online.map(lambda x: 100 if x == False else 0)
    res_df['R2'] = res_df.online.map(lambda x: 100 if x == False else 0)
    res_df['R1'] = res_df.tier.map(lambda x: 1400 if x == 1 else 0)
    res_df['R1'] = res_df.tier.map(lambda x: 1200 if x == 2 else 0) 
    res_df['R1'] = res_df.tier.map(lambda x: 1000 if x == 3 else 0)
    res_df['R1'] = res_df.tier.map(lambda x: 800 if x == 4 else 0)  
    res_df['R1'] = res_df.tier.map(lambda x: 600 if x == 5 else 0)
    res_df['R1'] = res_df.tier.map(lambda x: 400 if x == 6 else 0) 
    res_df['R1'] = res_df.tier.map(lambda x: 200 if x == 7 else 0)  
    res_df['R2'] = res_df.tier.map(lambda x: 1400 if x == 1 else 0)
    res_df['R2'] = res_df.tier.map(lambda x: 1200 if x == 2 else 0) 
    res_df['R2'] = res_df.tier.map(lambda x: 1000 if x == 3 else 0)
    res_df['R2'] = res_df.tier.map(lambda x: 800 if x == 4 else 0)  
    res_df['R2'] = res_df.tier.map(lambda x: 600 if x == 5 else 0)
    res_df['R2'] = res_df.tier.map(lambda x: 400 if x == 6 else 0) 
    res_df['R2'] = res_df.tier.map(lambda x: 200 if x == 7 else 0)  
    R_T_1 = R_T_1 + K * (res_df['R1'] - Pa) 
   
    R_T_2 = R_T_2 + K * ( res_df['R2'] - Pb) 
    
    
    df1['R_T_1'] = R_T_1
    df1['R_T_2'] = R_T_2
    df1['team_1'] = res_df['team_1']
    df1['team_2'] = res_df['team_2']
    df1['match_id'] = res_df['match_id']
    df1['Date'] = res_df['date']
    df1['Date'] = pd.to_datetime(df1['Date'])
    dfs = pd.DataFrame()
   
   
    #for name,  group in df1.drop_duplicates().groupby(['Date', 'match_id']):
        #print(name)
        #print(group.groupby(['match_id','team_1', 'team_2']).sum())
    
    t = df1.groupby(['match_id','team_1', 'team_2']).sum()
  
    s = t.groupby('team_1')['R_T_1'].sum()
   
    s2 = t.groupby('team_2')['R_T_2'].sum()
    dfs2 = pd.DataFrame(
        columns = ['Team','Ra'])
    dfs2 = s2
    result = pd.DataFrame()
    result = pd.concat((s, dfs2), axis=1, sort=True)
    result['R'] = (result['R_T_1'] + result['R_T_2'])/2
    result.fillna(0)
    result
    
    return result.sort_values(by='R', ascending=False)

        
    
R_T_1 = 100
R_T_2 = 100
RT1=0
RT2=0
K = 20
s = players_t
g = elo_rating(R_T_1, R_T_2, K, s)
g = g.drop(['R_T_1', 'R_T_2'], axis=1)
g = g.fillna(0)
g
```
